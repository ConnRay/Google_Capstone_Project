```python
# create distance column distance_km

clean_cyclstic["distance_km"] = ""

# this is a convoluted way to implement a nice simple function in a way that is time effective
tempdf = clean_cyclstic[['start_lat', 'start_long', 'end_lat', 'end_long']].copy()

tempdf['lat'] = tempdf['end_lat'] - tempdf['start_lat']
tempdf['lon'] = tempdf['end_long'] - tempdf['start_long']

tempdf['dlat'] = tempdf['lat'].astype(float).apply(math.radians)
tempdf['dlon'] = tempdf['lat'].astype(float).apply(math.radians)

tempdf['a'] = tempdf['dlat'].astype(float).div(2).apply(math.sin) * tempdf['dlat'].astype(float).div(2).apply(math.sin) + tempdf['start_lat'].astype(float).apply(math.radians).apply(math.cos) * tempdf['end_lat'].astype(float).apply(math.radians).apply(math.cos) * tempdf['dlon'].astype(float).div(2).apply(math.sin) * tempdf['dlon'].astype(float).div(2).apply(math.sin)
tempdf['one_minus_a'] = 1 - tempdf['a']
tempdf['y/x'] = tempdf['a'].astype(float).apply(math.sqrt) / tempdf['one_minus_a'].astype(float).apply(math.sqrt)
tempdf['c'] = tempdf['y/x'].astype(float).apply(math.atan)
radius = 6371
tempdf['d'] = radius * tempdf['c']
clean_cyclstic['distance_km'] = tempdf['d']
```

```python
# create a day of the week column called week_day
clean_cyclstic['week_day'] = clean_cyclstic['start_time'].dt.day_name()

# create a month coumn called month
clean_cyclstic['month'] = clean_cyclstic['start_time'].dt.month_name()
clean_cyclstic.head()
```

```python
# need to clean out some outliers from trip time which I'm going to say is any trip less than 1 minute or any trip over 1 day
# the outliers are probably indicative of someone changing their mind before the trip starts or someone forgetting to return their bike
clean_cyclstic = clean_cyclstic[clean_cyclstic['trip_time'] > pd.Timedelta(1,unit='min')]
clean_cyclstic = clean_cyclstic[clean_cyclstic['trip_time'] < pd.Timedelta(1,unit='d')]
clean_cyclstic.shape[0]
```

```python
# get mean, median on trip_time based on membership_type

trip_mean = clean_cyclstic.groupby('membership_type')['trip_time'].mean(numeric_only=False)
trip_median = clean_cyclstic.groupby('membership_type')['trip_time'].median(numeric_only=False)

print('mean: ', trip_mean)
print('median: ', trip_median)
```

```python
# get mean, median, max, min on distance_km based on membership_type

distance_mean = clean_cyclstic.groupby('membership_type')['distance_km'].mean()
distance_median = clean_cyclstic.groupby('membership_type')['distance_km'].median()
distance_max = clean_cyclstic.groupby('membership_type')['distance_km'].max()
distance_min = clean_cyclstic.groupby('membership_type')['distance_km'].min()

print('mean: ', distance_mean)
print('median: ', distance_median)
print('max: ', distance_max)
print('min: ', distance_min)
```

```python
# make a pie chart of bike types versues ridership type
import matplotlib.pyplot as plt

casual = clean_cyclstic[clean_cyclstic['membership_type'] == 'casual']
casual_electric = len(casual[casual['bike_type'] == 'electric_bike'])
casual_classic = len(casual[casual['bike_type'] == 'classic_bike'])
casual_docked = len(casual[casual['bike_type'] == 'docked_bike'])

member = clean_cyclstic[clean_cyclstic['membership_type'] == 'member']
member_electric = len(member[member['bike_type'] == 'electric_bike'])
member_classic = len(member[member['bike_type'] == 'classic_bike'])
member_docked = len(member[member['bike_type'] == 'docked_bike'])
```

```python
labels = 'Electric Bike', 'Classic Bike', 'Docked Bike'
sizesC = [casual_electric, casual_classic, casual_docked]
sizesM = [member_electric, member_classic, member_docked]

fig1, (ax1,ax2) = plt.subplots(1,2)
explode = (0, 0.1, 0)
ax1.pie(sizesC, explode = explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  

ax2.pie(sizesM,explode = explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax2.axis('equal')  

ax1.title.set_text('Casual Riders')
ax2.title.set_text('Membership Riders')

fig1.tight_layout()

plt.show()


```
